```python
movies = movies[ ['movie_id','title','overview','genres','keywords','cast','crew'] ]
movies.head(1)
```

# Data Preprocessing

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.drop_duplicates(inplace=True)
```

```python
#String looks like a Python list....convert to List
import ast 
```

```python
def convert(obj):
    L=[]
    lst=ast.literal_eval(obj)
    for d in lst:
        L.append( d["name"] )
    return L
```

```python
movies['genres'] = movies['genres'].apply( convert )
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
#Now for Cast , take only top3 cast...Otherwise very long
def convert3(obj):
    L=[]
    lst=ast.literal_eval(obj)
    count=0
    for d in lst:
        if count != 3:
            L.append( d["name"] )
            count+=1
        else:
            break
    return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
def fetch_director(obj):
    L=[]
    lst=ast.literal_eval(obj)
    for d in lst:
        if d["job"]=="Director":
            L.append( d["name"] )
            break
    return L
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
#Convert overview values from string to list..... str.split()
#lambda is a way to write a small, one-line function without giving it a name.

movies['overview']=movies['overview'].apply( lambda x : x.split() )
```

```python
movies.head()
```

```python
def remove_spaces(L):
    L1=[]
    for s in L:
        L1.append( s.replace(" ","") )
    return L1
```

```python
movies['genres'] = movies['genres'].apply( remove_spaces )
movies['keywords'] = movies['keywords'].apply( remove_spaces )
movies['cast'] = movies['cast'].apply( remove_spaces )
movies['crew'] = movies['crew'].apply( remove_spaces )
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
new_df=movies[ ['movie_id','title','tags'] ]
```

```python
new_df.head()
```

```python
#Tags...list to string...join by space " "
new_df["tags"]=new_df["tags"].apply( lambda x: " ".join(x) )

#Converted to string... now do in lower case
new_df["tags"]=new_df["tags"].apply( lambda x: x.lower() )
```

```python
new_df['tags'][0]  # 'overview','genres','keywords','cast'(Top 3 actors),'crew'(Director)
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    L=[]
    for word in text.split():
        L.append( ps.stem(word) )
    return " ".join(L)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
new_df.head()
```

# Vectorization

```python
'''Imports the tool used to convert text into numbers.
Creates a CountVectorizer object:
max_features=5000 ‚Üí keeps only the top 5000 most frequent words
stop_words='english' ‚Üí removes common words like the, is, and, of '''

from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector= cv.fit_transform( new_df['tags'] ).toarray()   #toarray() ‚Üí converts sparse matrix to a NumPy array
```

```python
vector.shape #4806 movies --> size of one vector(each row) 5000
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity( vector )
```

```python
similarity.shape
```

```python
new_df.head(1)
```

# Main function

```python
def recommend(movie):
    index=new_df[ new_df['title']==movie ].index[0]
    scores=similarity[index]
    movies_list= list( enumerate(scores) )
    movies_list= sorted( movies_list, key=lambda x: x[1], reverse=True )

    for i in movies_list[1:6]:
        print( new_df.iloc[ i[0] ].title )
```

```python
recommend('Batman Begins')
```
